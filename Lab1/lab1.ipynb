# Loading and Visuaizing Image 

## loading/reading image using OpenCv

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("IAU_Logo.jpeg")
print(img.shape)
plt.imshow(img)
plt.axis("off")
```

## Loding/reading inage using PIL

```python
from PIL import Image
import matplotlib.cm as cm
#
imag2 = Image.open("KFUPM_logo.png")
print(imag2.size)
plt.axis("off")
plt.imshow(imag2)
```

# Image Storing

## a.	Saving image to disc using OpenCV


```python
cv2.imwrite("Not_KFUPM_logo.png", img)
```

##  b.	Saving image to disc using PIL

```python
imag2.save("Not_IAU_logo.png")
```

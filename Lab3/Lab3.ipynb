* Image Shearing
* * Shear mapping is a linear map that displaces each point in a fixed direction, it substitutes every point horizontally or vertically by a specific value in proportion to its x or y coordinates, there are two types of shearing effects.

* Shearing in the x-axis Direction
* * When shearing is done in the x-axis direction, the boundaries of the image that are parallel to the x-axis keep their location, and the edges parallel to the y-axis change their place depending on the shearing factor:


https://thepythoncode.com/article/image-transformations-using-opencv-in-python

```python
import matplotlib.pyplot as plt
import numpy as np

img_BGR2RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img_BGR2RGB)
rows, cols, dim = img.shape
# transformation matrix for Shearing
# shearing applied to x-axis
M = np.float32([[1, 1, 0],
             	[0.1, 1  , 1],
            	[0,0  , 1]])
sheared_img = cv2.warpPerspective(img_BGR2RGB,M,(int(cols*1.5),int(rows*1.5)))
plt.axis('off')
plt.imshow(sheared_img)
plt.show()
```

```python
negative_image = 256 - 1 - img 
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.title("Original image")
plt.axis('off')
plt.imshow(img_BGR2RGB)
plt.subplot(1,2,2)
plt.title("Negative image")
plt.imshow(negative_image)
plt.axis('off')
plt.show()
```

```python
c = 45
log_transformed = c * (np.log(img + 1))
# Convert to 8-bit unsigned integer format
log_transformed = np.uint8(log_transformed)
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.title("Original image")
plt.axis('off')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title("Log transformed image")
plt.axis('off')
plt.imshow(log_transformed)
```

```python
gamma = 2
gamma2=4
adjusted_image = np.power(img / 255.0, gamma) * 255.0
adjusted_image = adjusted_image.astype(np.uint8)
adjusted_image2 = np.power(img / 255.0, gamma2) * 255.0
adjusted_image2 = adjusted_image2.astype(np.uint8)
plt.figure(figsize=(10, 5)) 
plt.subplot(1, 3, 2)  
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.title("adjusted_image gamma =2")

plt.subplot(1, 3, 1)  
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 3, 3)  
plt.imshow(cv2.cvtColor(adjusted_image2, cv2.COLOR_BGR2RGB))
plt.title("Original Image gamma = 4")
plt.axis('off')
```
